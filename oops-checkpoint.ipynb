```python
class Product():
    def __init__(self, id, marked_price, discount):
        self.id = id
        self.marked_price = marked_price
        self.discount = discount
#  
# create a read only property named selling_price that is calculated by
# deducting discount from the marked_price. 
    @property
    def selling_price(self):
        return self.marked_price - 0.01*self.discount*self.marked_price
        
    def display(self):
        print(self.id, self.marked_price, self.discount)
        
#  give an additional 2% discount on a product, if its price is above 500
    @property
    def discount(self):
        if self.marked_price > 500:
            return self.discount+2
        else:
            return self.discount
        
    @discount.setter
    def discount(self, new_discount):
        self.discount = new_discount
        
p1 = Product('X879', 400, 6)
p2 = Product('A234', 100, 5)
p3 = Product('B987', 990, 4)
p4 = Product('H456', 800, 6)
print(p1.id, p1.selling_price)
print(p2.id, p2.selling_price)
print(p3.id, p3.selling_price)
print(p4.id, p4.selling_price)

p4.discount = 10
print(p4.id, p4.selling_price)
```

```python
# Write a Circle class with an instance variable named radius and a method named area. 
class Circle():
    def __init__(self, radius=0):
        self.radius = radius
        
    @property
    def radius(self):
        return self.radius
        
    @radius.setter
    def radius(self, new_radius):
        if new_radius > 0:
            self.radius = new_radius
        else:
            raise ValueError('Radius should be positive')
    
    def area(self):
        print(3.14*self.radius*self.radius) 
    
# Create two more attributes named diameter and circumference and make them behave as read only attributes. 

    @property
    def diameter(self):
        print(2*self.radius)
        
    @property
    def circumference(self):
        print(2*3.14*self.radius)
        
c1 = Circle(2)
c1.area()
c1.diameter
c1.circumference

c2 = Circle(3)
c2.area()
c2.diameter
c2.circumference
```
