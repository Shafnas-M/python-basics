```python
import tkinter as tk
window=tk.Tk()

window.geometry("510x600")
window.resizable(False, False)

widgets =['label', 'entry', 'password', 'radio buttons', 'check buttons', 'button']
x=0
for widget in widgets:
    label = tk.Label(window, text=widget)
    label.grid(row=x, column=0)
    x+=1

label = tk.Label(window, text="Label")
label.grid(row=0, column=1)

entry = tk.Entry(window)
entry.grid(row=1, column=1)

entry = tk.Entry(window, show="*")
entry.grid(row=2, column=1)

radio_button_frame=tk.Frame(window)
radio_button_frame.grid(row=3, column=1)
radiobutton1 = tk.Radiobutton(radio_button_frame, text="Radiobutton 1", value=1)
radiobutton2 = tk.Radiobutton(radio_button_frame, text="Radiobutton 2", value=2)
radiobutton1.pack()
radiobutton2.pack()

check_button_frame = tk.Frame(window)
check_button_frame.grid(row=4, column=1)
checkbutton1 = tk.Checkbutton(check_button_frame, text="Checkbutton 1",)
checkbutton2 = tk.Checkbutton(check_button_frame, text="Checkbutton 2",)
checkbutton1.pack()
checkbutton2.pack()

button = tk.Button(window, text="Submit",)
button.grid(rows=5, column=1)

window.mainloop()
```

```python
import tkinter as tk
window = tk.Tk()

window.geometry("510x600")
window.resizable(False, False)

global label
label = tk.Label(window, text="Please ensure that you use the correct data type", fg='red')

def submit():
    try:
        print(f'String Entry: {str_entry_var.get()}')
        print(f'Integer Entry: {int_entry_var.get()}')
        print(f'Double Entry: {double_entry_var.get()}')
        print(f'Boolean Entry: {boolean_entry_var.get()}')
    except:
        label.pack()
    else:
        label.pack_forget()

str_entry_var = tk.StringVar()
str_entry=tk.Entry(window, textvariable=str_entry_var)
str_entry.pack()

int_entry_var = tk.IntVar()
int_entry = tk.Entry(window, textvariable = int_entry_var)
int_entry.pack()

double_entry_var = tk.DoubleVar()
double_entry = tk.Entry(window, textvariable = double_entry_var)
double_entry.pack()

boolean_entry_var = tk.BooleanVar()
boolean_entry = tk.Entry(window, textvariable = boolean_entry_var)
boolean_entry.pack()

button = tk.Button(window, text="Submit", command=submit)
button.pack()

window.mainloop()
```
